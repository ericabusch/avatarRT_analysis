```python
from scipy.ndimage import gaussian_filter1d

# Smooth the lines using a rolling window
time = np.arange(0, 100, 1)
line1 = np.sin(time/10) + np.random.normal(0, 0.5, size=time.shape)
line2 = np.sin(time/10 + np.pi/4) + np.random.normal(0, 0.5, size=time.shape)
line3 = -np.sin(time/10) + np.random.normal(0, 0.5, size=time.shape)
line4 = np.random.normal(0, 1, size=time.shape) 
# Apply Gaussian smoothing to each line
sigma = 2  # smoothing parameter
line1_smooth = gaussian_filter1d(line1, sigma=sigma)
line2_smooth = gaussian_filter1d(line2, sigma=sigma)
line3_smooth = gaussian_filter1d(line3, sigma=sigma)
line4_smooth = gaussian_filter1d(line4, sigma=sigma)
# Normalize everything to zero mean and unit variance
line1_smooth = zscore(line1_smooth)
line2_smooth = zscore(line2_smooth)
line3_smooth = zscore(line3_smooth)
line4_smooth = zscore(line4_smooth)
# Create presentation-ready plot
fig, ax = plt.subplots(1, 1, figsize=(8, 5))

# Plot smoothed lines with better styling
ax.plot(time, line1_smooth, label='Sadness', color='orange', linewidth=2.5, alpha=0.9)
ax.plot(time, line2_smooth, label='Disappointment', color='red', linewidth=2.5, alpha=0.9)
ax.plot(time, line3_smooth, label='Joy', color='pink', linewidth=2.5, alpha=0.9)
ax.plot(time, line4_smooth, label='Disgust', color='purple', linewidth=2.5, alpha=0.9)

# Styling
# TUrn off x and y ticks
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Time in Movie', fontsize=14, fontweight='bold')
ax.set_ylabel('Intensity', fontsize=14, fontweight='bold')
ax.legend(title='Category',loc='best', frameon=True, fancybox=True, shadow=False, fontsize=11)
# ax.grid(True, alpha=0.3, linestyle='--', linewidth=0.5)
sns.despine()

plt.tight_layout()
# plt.savefig('plots/timeseries_smoothed.pdf', transparent=True, bbox_inches='tight', dpi=300)
plt.show()
```
